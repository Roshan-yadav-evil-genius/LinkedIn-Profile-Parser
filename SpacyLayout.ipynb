```python
import spacy
from spacy_layout import spaCyLayout
from rich.console import Console
from rich.markdown import Markdown
from rich import print,print_json
from spacy.tokens import Span
from typing import List
import json
```

```python
from labeler import set_extensions, label_linkedin_headers, label_timelines,label_locations,label_experience_jobroles,label_experience_organizations,label_education_organizations,linkedin_header_to_category_value,get_dict_output

from config import SECTION_HEADERS, LinkedInCategory, LinkedInLabels
```

```python
nlp = spacy.load("en_core_web_sm")
layout = spaCyLayout(nlp)
console = Console()
```

```python
set_extensions()
```

```python
doc = layout("./data/Ceren Kaya1.pdf")
```

```python
def is_valid_url(token)->bool:
    return token.like_url or any(domain in token.text for domain in SECTION_HEADERS)
```

```python
spans=doc.spans["layout"]
new_span_phase1:List[Span]=label_linkedin_headers(spans)
new_span_phase2:List[Span]=label_timelines(new_span_phase1)
new_span_phase3:List[Span]=label_locations(new_span_phase2,nlp)
new_span_phase4:List[Span]=label_experience_jobroles(new_span_phase3)
new_span_phase5:List[Span]=label_experience_organizations(new_span_phase4)
new_span_phase6:List[Span]=label_education_organizations(new_span_phase5,nlp)
```

```python
content=""
first_userinfo=True
for span in new_span_phase6:
    info_str = "\n" 
    if span._.linkedin_label == "section_header" and span._.linkedin_category==linkedin_header_to_category_value(span.text):
        info_str += f"[bold red on white]{span.text}[/]"
    else:
        if span._.linkedin_category == "userInfo" and first_userinfo:
            info_str += f"[bold red on white]userinfo[/]\n"
            first_userinfo = False
        info_str += (
            f"Label: [bold green]{span._.linkedin_label}[/] "
            f"Category: [bold cyan]{span._.linkedin_category}[/] "
            f"Text: [yellow]{span.text}[/] "
        )
    
    content += info_str
print(content)
```
